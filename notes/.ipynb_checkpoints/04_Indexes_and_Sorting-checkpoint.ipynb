```python
countries = pd.read_csv('../data/world-happiness-report-2021.csv', index_col='Country name')
# we can also use the positional index of the column, eg. index_col=0
countries['Healthy life expectancy'].nlargest()
```

```python
houses = pd.read_csv('../data/kc_house_data.csv')
titanic = pd.read_csv('../data/titanic.csv')
```

```python
houses.sort_values('price', ascending=False)
```

```python
houses.sort_values(['bedrooms', 'bathrooms'], ascending=False) 
# first sorts by bedrooms, then bathrooms if bedroom is the same
```

```python
countries.loc['Denmark':'France'] # using slices with sorted text indices
```

## iloc, integer location

```python
countries.iloc[1] # single row turned into series
```

```python
countries.iloc[[0, 10, 20]] # multiple rows by index
```

```python
countries.iloc[0:10] # multiple rows by slice
```

```python
houses.sort_index(ascending=False, inplace=True)
```

```python
houses.iloc[[1,2]] # notice we are getting the first 10 after ordering descending
```

```python
houses.loc[21611:21610, ['price', 'bedrooms']]
```

```python
titanic.loc[50:60:2, ['name', 'sex', 'age']]
```

```python
countries.loc['Canada':'Denmark', 'Generosity'] # return a series
```

```python
houses['bedrooms'].value_counts().iloc[0:3]
```

```python
titanic['age'].value_counts().loc['18']
```
